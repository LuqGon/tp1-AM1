## __Punto 1__ ##

### __Enunciado__ ###
Hallar los puntos de intersección de los gráficos de f(x) y g(x), siendo:

$$f(x) = (x − 3)^{-2}$$

$$g(x) = \frac{1024}{(x − 3)^{-3}}$$


### __Desarrollo__ ###

Para encontrar los puntos de intersección, igualamos $f(x)$ y $g(x)$:

$$(x − 3)^{-2} = \frac{1024}{(x − 3)^{-3}}$$

Resolviendo esta ecuación, determinamos los valores de $x$ donde los gráficos de $f(x)$ y $g(x)$ se intersectan.

$$\frac{1}{1024} = \frac{1}{(x − 3)^{-3}(x − 3)^{-2}}$$
$$\frac{1}{1024} = \frac{1}{(x − 3)^{-5}}$$
$$\frac{1}{1024} = (x − 3)^{5}$$
$$\frac{\sqrt[5]{1}}{\sqrt[5]{1024}} = x − 3$$
$$\frac{1}{4}+{3} = x$$
$$\frac{13}{4} = x$$

En la siguiente celda se corroborará si el resultado obtenido es el correcto, teniendo en cuenta que:
$$f(\frac{13}{4}) = g(\frac{13}{4})$$

```python
# Definimos f(x) y g(x)
f1 = lambda x: (x-3)**(-2)
g1 = lambda x: 1024/(x-3)**(-3)
if(f1(13/4) == g1(13/4)):
    print(f"El punto de intersección obtenido es correcto")
else:
    print("El punto de intersección obtenido es incorrecto")
print(f'f(13/4)={f1(13/4)}')
print(f'g(13/4)={g1(13/4)}')
```

Se procede a graficar las funciones

```python
#Grafico f y g
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 1000)
x1 = np.linspace(0, 3, 1000)
x2 = np.linspace(3, 5, 1000)

fig = plt.figure(figsize=(10, 10))
plt.plot(x1, f1(x1), label='f(x)', color='green')
plt.plot(x2, f1(x2), color='green')
plt.plot(x, g1(x), label='g(x)', color='magenta')  
plt.scatter(13/4, 16, color='black', label='Punto de intersección',zorder=2)  
plt.text(13/4, 16, '(3.25, 16)', fontsize=9, verticalalignment='bottom', horizontalalignment='left')
plt.axhline(0, color='black', ls='-')

plt.title('Grafico de la interseccion entre f(x) y g(x)')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend(loc='lower left')
plt.xlim(0,5)
plt.ylim(-4,18)
plt.grid(visible=True, which='both', color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
