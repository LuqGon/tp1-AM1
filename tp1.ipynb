## Enunciado ##
Hallar los puntos de intersección de los gráficos de f(x) y g(x), siendo:

$$f(x) = (x − 3)^{-2}$$

$$g(x) = \frac{1024}{(x − 3)^{-3}}$$


### Desarrollo ###

Para encontrar los puntos de intersección, igualamos $f(x)$ y $g(x)$:

$$(x − 3)^{-2} = \frac{1024}{(x − 3)^{-3}}$$

Resolviendo esta ecuación, determinamos los valores de $x$ donde los gráficos de $f(x)$ y $g(x)$ se intersectan.

$$\frac{1}{1024} = \frac{1}{(x − 3)^{-3}(x − 3)^{-2}}$$
$$\frac{1}{1024} = \frac{1}{(x − 3)^{-5}}$$
$$\frac{1}{1024} = (x − 3)^{5}$$
$$\frac{\sqrt[5]{1}}{\sqrt[5]{1024}} = x − 3$$
$$\frac{1}{4}+{3} = x$$
$$\frac{13}{4} = x$$

En la siguiente celda se corroborará si el resultado obtenido es el correcto, teniendo en cuenta que:
$$f(\frac{13}{4}) = g(\frac{13}{4})$$

```python
# Definimos f(x) y g(x)
f = lambda x: (x-3)**(-2)
g = lambda x: 1024/(x-3)**(-3)
if(f(13/4) == g(13/4)):
    print(f"El punto de intersección obtenido es correcto")
else:
    print("El punto de intersección obtenido es incorrecto")
print(f'f(13/4)={f(13/4)}')
print(f'g(13/4)={g(13/4)}')
```

Se procede a graficar las funciones

```python
#Grafico f y g
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2.5, 3.3, 100)
x1 = np.linspace(2.5, 2.95, 50)
x2 = np.linspace(3.05, 3.3, 50)

fig = plt.figure(figsize=(6, 6))
plt.plot(x1, f(x1), label='f(x)', color='green')
plt.plot(x2, f(x2), color='green')
plt.plot(x, g(x), label='g(x)', color='pink')  
plt.scatter(13/4, 16, color='black', label='Punto de intersección',zorder=2)  
plt.text(13/4, 16, '(13/4, 16)', fontsize=9, verticalalignment='bottom', horizontalalignment='left')
plt.title('Gráfica de interseccion entre f(x) y g(x)')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend(loc='lower right')
plt.show()
```

## Enunciado ## 

Elaborar un programa en Python que verique si x = 3 y x = 9 son soluciones de
la ecuación:

$$(x-6)^{2} = x^{2}$$


¿Hay otros valores de x que satisfacen la ecuación? Justicarlo.


### Desarrollo ###

Para verificar si x = 3 y x = 9 son soluciones de f(x), se debe que cumplir que:

$$f(3) = 0 \quad \text{y} \quad f(9) = 0$$

```python
# Defino una funcion para evaluar si los valores son soluciones de la funcion matematica f(x)
def son_soluciones(posibles_soluciones):
    for x in posibles_soluciones:
        if(f(x) == 0):
            print(f"{x} es solucion de la ecuacion f(x)")
        else:
            print(f"{x} no es solucion de la ecuacion f(x)")

# Defino la funcion matematica f(x) que se nos dio en el problema
f= lambda x : x**2-(x-6)**2

# Evaluo las posibles soluciones que se nos dio en el enunciado
posibles_soluciones = [3,9]
son_soluciones(posibles_soluciones)
```

El valor que es solucion es x = 3

Continuamos con la segunda parte del enunciado: 

Para averiguar si hay otra solucion, debemos desarrollar la ecuacion aunque podemos inferir que solo hay una solucion que deberia ser x = 3
$$(x − 6)^{2} = x^{2}$$
$$(x − 6)(x - 6) = x^{2}$$
$$x^{2}-12x+36 = x^{2}$$
$$\cancel{x^{2}}-12x+36 = \cancel{x^{2}}$$
$$-12x=-36$$
$$x=\frac{-36}{-12}$$
$$x=3$$
De esta forma de demuestra y concluye con que solo hay una posible solucion
