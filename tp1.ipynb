## Hallar los puntos de intersección de los gráficos de f(x) y g(x)

Dadas las funciones:

$$f(x) = (x − 3)^{-2}$$

$$g(x) = \frac{1024}{(x − 3)^{-3}}$$

Para encontrar los puntos de intersección, igualamos $f(x)$ y $g(x)$:

$$(x − 3)^{-2} = \frac{1024}{(x − 3)^{-3}}$$

Resolviendo esta ecuación, determinamos los valores de $x$ donde los gráficos de $f(x)$ y $g(x)$ se intersectan.

$$\frac{1}{1024} = \frac{1}{(x − 3)^{-3}(x − 3)^{-2}}$$
$$\frac{1}{1024} = \frac{1}{(x − 3)^{-5}}$$
$$\frac{1}{1024} = (x − 3)^{5}$$
$$\frac{\sqrt[5]{1}}{\sqrt[5]{1024}} = x − 3$$
$$\frac{1}{4}+{3} = x$$
$$\frac{13}{4} = x$$




```python

x = 13/4
f = (x-3)**(-2)
g = 1024/(x-3)**(-3)
if(f == g):
    print("El punto de intersección obtenido es correcto")
else:
    print("El punto de intersección obtenido es incorrecto")
print(f'f(x)={f}, g(x)={g}')
```

```python
#Grafico f y g
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2, 4, 100)
x1 = np.linspace(2, 3, 100)
x2 = np.linspace(3, 4, 100)
f1 = (x1-3)**(-2)
f2 = (x2-3)**(-2)
g = 1024/(x-3)**(-3)

fig = plt.figure(figsize=(6, 6))
plt.plot(x1, f1, label='f(x)', color='blue')
plt.plot(x2, f2, label='f(x)', color='blue')
plt.plot(x, g, label='g(x)', color='orange')  
plt.scatter(13/4, 16, color='red', label='Intersección')  
plt.title('Gráfica de f(x) y g(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(3, color='red', lw=1, ls='--',label='Asintota vertical')
plt.legend()
plt.show()
```
